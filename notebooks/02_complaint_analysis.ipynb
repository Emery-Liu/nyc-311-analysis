```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/nyc_311_clean.csv")
```

**Top Complaint Types**

```python
# most common complaint types with bar chart
complaints = df["complaint_type"].value_counts().head(10).sort_values()
display(complaints)
complaints.plot.barh()

plt.title("Top 10 Complaint Types")
plt.xlabel("Complaint Type")
plt.ylabel("Number of Complaints")

plt.show()
```

**Borough Distribution**

```python
# "Complaint Distribution by Borough
reg_complaint_count = df["borough"].value_counts().head(10)
display(reg_complaint_count)
reg_complaint_count.plot.pie()

plt.title("Complaint Distribution by Borough")
plt.show()
```

**Agency Distribution**

```python
act_agency = df["agency"].value_counts().head(10)
display(act_agency)
act_agency.plot.bar()

plt.title("Top 10 activated agency Types")
plt.xlabel("Agency Type")
plt.ylabel("Number of activate")

plt.show()
```

```python
# top five complaint types by borough
top5_by_borough = (
    df.groupby("borough")["complaint_type"]
      .value_counts()
      .groupby(level=0)
      .head(5)
)
display(top5_by_borough)
```
