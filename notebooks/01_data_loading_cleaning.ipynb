```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/nyc_311_2025_sample.csv")
```

```python
df.head()
```

```python
# shape of the dataframe
df.shape
```

```python
# general information of the dataframe
df.info()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# convert string type to datetime type
df["created_date"] = pd.to_datetime(df["created_date"])
df["closed_date"] = pd.to_datetime(df["closed_date"])
```

```python
# Average resolution time
df["resolution_time"] = df["closed_date"] - df["created_date"]
df["resolution_hours"] = df["resolution_time"].dt.total_seconds() / 3600
```

```python
df.to_csv(
    "../data/processed/nyc_311_clean.csv",
    index=False
)
```
