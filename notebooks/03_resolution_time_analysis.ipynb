```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/nyc_311_clean.csv")
```

**Average Resolution Time**

```python
avg_hours = np.mean(df["resolution_hours"].dropna())
print("the average resolution time is: {} hours".format(avg_hours))
```

**Median Resolution Time**

```python
median_restime = np.median(df["resolution_hours"].dropna())
print("the median resolution time is: {} hours".format(median_restime))
```

**95 Percentile of Resolution Time**

```python
Nine5_percentile = np.percentile(df["resolution_hours"].dropna(), 95)
print("The 95 percentile of resolution time is {}".format(Nine5_percentile))
```

**Average Resolution Time by Agency**

```python
agency_restime_avghour = df.groupby("agency")["resolution_hours"].mean()
agency_restime_avghour.sort_values().plot.barh(color = "skyblue")

plt.title("Average Resolution Time(hours) by Agency")
plt.ylabel("Hours")
plt.show()
```

**#The top ten complain types with slow average resolution time**

```python
slow_complaint_time = df.groupby("complaint_type")["resolution_hours"].mean().sort_values(ascending=False).head(10)
slow_complaint_time.plot.barh(color="steelblue")

plt.xlabel("Average Resolution Time (Hours)")
plt.ylabel("Complaint Type")
plt.title("Top 10 Complaint Types with Longest Resolution Time")

plt.show()
```
