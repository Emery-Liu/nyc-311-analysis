```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/nyc_311_clean.csv")
```

```python
clean_df = df.dropna(
    subset=["latitude", "longitude"]
)

plt.figure(figsize=(10,8))

plt.hist2d(
    clean_df["longitude"],
    clean_df["latitude"],
    bins=120
)

plt.colorbar(label="Complaint Count")

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.title(
    "NYC 311 Complaint Density Heatmap"
)

plt.show()
```
